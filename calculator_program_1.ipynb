```python
# menu driven calculator
# take the input from user
num1=int(input("enter the first number:"))
num2=int(input("enter the second number:"))
operation=input("enter the operation:")
if operation=="+":
    print(num1+num2)
elif operation=="-":
    print(num1-num2)
elif operation=="*":
    print(num1*num2)
elif operation=="/":
    print(num1/num2)
elif operation=="%":
    print(num1%num2)
elif operation=="**":
    print(num1**num2)
else:
    print("invalid operation")
```
