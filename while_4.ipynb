```python
# guess number
guess=int(input("guess number:"))
original_num=5
while guess != original_num:
    if guess < original_num:
      print("smallest number")
      print("please try again")
    else:
      print("biggest number")
      print("please try again")
    guess=int(input("guess number:"))
print("correct number")
```
